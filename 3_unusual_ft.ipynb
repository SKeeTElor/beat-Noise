```python
import numpy as np
import matplotlib.pyplot as plt

tf = .99
N= 100
f = 40


t = np.linspace(0, tf, N)

xf = np.linspace(-N/2/tf, N/2/tf, N)


s = np.cos(2*np.pi*f*t)

ft = np.fft.fftshift(np.fft.fft(s))


plt.plot(xf, ft)

```
